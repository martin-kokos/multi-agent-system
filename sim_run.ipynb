## Run simulation

```python
from main import Environment

from bokeh.io import output_notebook
import bokeh.plotting as plt
import numpy as np

import logging
from pprint import pprint

#logging.basicConfig(level=logging.INFO)

output_notebook()
```

```python
config = {
 'seed': 0,
 'max_agents': 200,

'agent_confidence_gain': 0.1,
 'agent_convo_disagreement': 0.25,
 'agent_evolve_rate': 0.5,
 'agent_poly_degree': 5,
 'gov_term': 50,
 'issue_solving_disagreement': 0.05,
 'num_issues': 5}



env = Environment(**config)

ticks = 10_000

env.run(ticks=ticks)
```

## Detailed report

```python
summary = env.stats.summary_metrics()
steady_solve_auc = summary['elapsed_time'] * summary['issues_solved'] / 2
auc_improvement = 2 * summary['issues_solved_auc'] / (steady_solve_auc + 0.0001)  # improvement/t
reasonable_solved = (1000 - summary['issues_solved']) / 1000  # solved num
poly_stability = np.log(summary['poly_complexity_var'])  # pop stability

print(f'{auc_improvement=}')
print(f'{reasonable_solved=}')
print(f'{poly_stability=}')
```

```python
env.stats.plot('agent:')
env.stats.plot('agents.mean_confidence')
env.stats.plot('agents.mean_poly_complexity')
env.stats.plot('gov.num_agents')
env.stats.plot('issues.solved')
plt.show(env.stats.plot_personalities(env))
```
